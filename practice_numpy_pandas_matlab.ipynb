```python
X = np.array([
    [2, 3],
    [2, 5],
    [3, 4],
    [5, 9],
])
```

```python
ones = np.ones((X.shape[0], 1))
```

```python
X = np.concatenate((ones,X), axis=1)
```

```python
t = np.array([1,5,6,8])
t
```

```python
xx = np.dot(X.T, X)
```

```python
xx_inv = np.linalg.inv(xx)

xx_inv
```

```python
np.dot(xx, xx_inv)
```

```python
xt = X.T.dot(t)

w = np.dot(xx_inv, xt)
w
```

```python
w_ = np.linalg.solve(X.T.dot(X), X.T.dot(t))
w_
```

np.linalg.solveは、連立一次方程式を解く

A X = B

変形

X = invA B

# Scikit-learn

```python

from sklearn.datasets import load_boston
dataset = load_boston()

x= dataset.data
t= dataset.target

x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, t_train, t_test = train_test_split(x,t,test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()
```

```python
reg_model.fit(x_train, t_train)
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
reg_model.score(x_train, t_train)
```

```python
reg_model.predict(x_test[0:2])
```

```python
reg_model.score(x_test, t_test)
```

```python
from sklearn.preprocessing import StandardScaler, PowerTransformer

# scaler = StandardScaler()
scaler = PowerTransformer()


scaler.fit(x_train)
```

```python
scaler.mean_
```

```python
scaler.var_
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

reg_model1 = LinearRegression()
reg_model1.fit(x_train_scaled, t_train)
```

```python
reg_model1.score(x_train_scaled, t_train)
```

```python
reg_model1.score(x_test_scaled, t_test)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('scaler', PowerTransformer()),
    ('reg', LinearRegression())
])
```

```python
pipeline.fit(x_train, t_train)
```

```python
pipeline.score(x_train, t_train)
```

```python
linear_result = pipeline.score(x_test, t_test)
linear_result
```
